```python
# We are importing requests over here for fetching data from the Api
# pandas is used to read the API
import requests
import pandas as pd
```

```python
# Here created a Variable called response in which we are storing the data fetched from api
# .json function is used to to read the data in the form of json
response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1').json()
response
```

```python
# creating Dictionary
d = {'Name': ["Omkar",'Jitesh'], 'Age': [45,28]}
```

```python
# Stored Dictionary in Dataframe
df = pd.DataFrame(d)
df
```

```python
# Created empty list in which we will be going to append the data present in it 
popularity = []
vote_count = []
original_language = []
title = []
vote_average = []
overview = []
release_date = []

# Making use of Loop on  response Variable where in empty list we are appending data present in json in particular field


for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
# Over here we are creating a dictionary in which we want the particular data from api which we are going to Store
d = {"title":title , 'overview': overview ,'original_language' : original_language, 'release_date':release_date,'popularity':popularity,'vote_count':vote_count,'vote_average':vote_average}
# The dictionary created will be store in Data frame
df = pd.DataFrame(d)
# Now this will read only a  single Page of a API
df
```

```python
# Here created a empty DataFrame with final Keyword
final = pd.DataFrame()

# Here we have created j varaible in which range function will be going to work from particular page to a particular page
# and over here we are making use of String format method to change the value of j
for j in range(1,50):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}.format{j}').json()
    popularity = []
    vote_count = []
    original_language = []
    title = []
    vote_average = []
    overview = []
    release_date = []
# Making use of Loop for response Variable
    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])

    d = {"title":title , 'overview': overview ,'original_language' : original_language, 'release_date':release_date,'popularity':popularity,'vote_count':vote_count,'vote_average':vote_average}

    df = pd.DataFrame(d)
# Over here in final Dataframe we are appending  various Dataframe from particular page to particular Page 
# ignore index = True which helps us for ignoring the sr no and putting sr no in Ascending Order
    final = final.append(df,ignore_index = True)
    
```
